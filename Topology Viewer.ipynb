```python
import pylab as p
import networkx as nx
from networkx.readwrite import json_graph
import json
import sys
```

Function for getting raw **JSON** and convert it to **Graph Object**

```python
def read_json_file_to_graph(filename):
    with open(filename) as f:
        js_graph = json.load(f)
    return json_graph.node_link_graph(js_graph)
```

Separating **switches** and **hosts** and assigning **Colors** to them

```python
with open('./topologies/simpleNet.json', 'r') as fd:
    jsonData = fd.read()
nodes = json.loads(jsonData)['nodes']
colors = [("blue" if X['id'][0] == "S" else "green") for X in nodes]
```

Drawing **Nodes** and **Links** to ***Visualize*** the whole **network**<br/>
- Create a JSON file (e.g. `'simpleNet.json'`)
- Define your topology in JSON network format
- Change the below path
- View your network

```python
nx.draw(read_json_file_to_graph('./topologies/simpleNet.json'), 
        with_labels=True, 
        font_color="yellow", 
        font_weight="bold", 
        node_size=700, 
        node_color=colors)
```
